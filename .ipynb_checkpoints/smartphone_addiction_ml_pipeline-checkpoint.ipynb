This notebook demonstrates a complete machine learning pipeline for smartphone addiction detection. The project includes:

- `addicted_label`: Binary classification (0 = Not Addicted, 1 = Addicted)
- `addiction_level`: Categorical (None, Mild, Moderate, Severe)

## 3. Data Cleaning and Preprocessing

### Key Data Quality Issues Found:
1. **Missing values** across 8 columns (2-5% each)
2. **Duplicate records** (~1.5% of data)
3. **Outliers and unrealistic values** (negative ages, screen time >24h, negative sleep, etc.)
4. **Inconsistent categorical values** (typos, case variations in gender, stress_level, etc.)
5. **Label inconsistencies** (addicted_label doesn't match addiction_level)
6. **Data type issues** (some numeric values stored as strings)

### 3.1 Handling Missing Values

```python
# Create a copy for cleaning
df_clean = df.copy()
print(f"Starting with {df_clean.shape[0]} records")

# ============================================================
# STEP 1: Remove Exact Duplicates
# ============================================================
print("\n--- Step 1: Remove Duplicate Records ---")
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
after = len(df_clean)
print(f"Removed {before - after} duplicate records")
print(f"Remaining: {after} records")

# ============================================================
# STEP 2: Fix Data Type Issues
# ============================================================
print("\n--- Step 2: Fix Data Type Issues ---")
if 'age' in df_clean.columns:
    df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')
    print(f"Converted 'age' to numeric")

for col in ['daily_screen_time_hours', 'social_media_hours', 'gaming_hours',
           'work_study_hours', 'sleep_hours', 'notifications_per_day',
           'app_opens_per_day', 'weekend_screen_time']:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
print("All numeric columns converted to proper types")

# ============================================================
# STEP 3: Handle Unrealistic Values and Outliers
# ============================================================
print("\n--- Step 3: Handle Unrealistic Values ---")

# Fix age: keep 13-100 range
age_before = df_clean['age'].notna().sum()
df_clean.loc[(df_clean['age'] < 13) | (df_clean['age'] > 100), 'age'] = np.nan
age_after = df_clean['age'].notna().sum()
print(f"Age: Set {age_before - age_after} unrealistic values to NaN (kept 13-100)")

# Fix screen time >24h
screen_over24 = (df_clean['daily_screen_time_hours'] > 24).sum()
df_clean.loc[df_clean['daily_screen_time_hours'] > 24, 'daily_screen_time_hours'] = np.nan
print(f"Daily screen time: Set {screen_over24} values >24h to NaN")

weekend_over24 = (df_clean['weekend_screen_time'] > 24).sum()
df_clean.loc[df_clean['weekend_screen_time'] > 24, 'weekend_screen_time'] = np.nan
print(f"Weekend screen time: Set {weekend_over24} values >24h to NaN")

# Fix negative numeric values
for col in ['social_media_hours', 'gaming_hours', 'work_study_hours',
           'notifications_per_day', 'app_opens_per_day']:
    if col in df_clean.columns:
        neg_count = (df_clean[col] < 0).sum()
        if neg_count > 0:
            df_clean.loc[df_clean[col] < 0, col] = np.nan
            print(f"{col}: Set {neg_count} negative values to NaN")

# Fix sleep hours: remove negative and >24h
sleep_issues = ((df_clean['sleep_hours'] < 0) | (df_clean['sleep_hours'] > 24)).sum()
df_clean.loc[(df_clean['sleep_hours'] < 0) | (df_clean['sleep_hours'] > 24), 'sleep_hours'] = np.nan
print(f"Sleep hours: Set {sleep_issues} unrealistic values to NaN")

# ============================================================
# STEP 4: Standardize Categorical Values
# ============================================================
print("\n--- Step 4: Standardize Categorical Values ---")

gender_mapping = {
    'male': 'Male', 'MALE': 'Male', 'M': 'Male', 'm': 'Male',
    'female': 'Female', 'FEMALE': 'Female', 'F': 'Female', 'f': 'Female',
    'other': 'Other', 'OTHER': 'Other', 'O': 'Other', 'o': 'Other'
}
if 'gender' in df_clean.columns:
    df_clean['gender'] = df_clean['gender'].map(gender_mapping).fillna(df_clean['gender'])
    print(f"Gender standardized: {df_clean['gender'].unique()}")

stress_mapping = {
    'low': 'Low', 'LOW': 'Low', 'L': 'Low', 'lo': 'Low',
    'medium': 'Medium', 'MED': 'Medium', 'Med': 'Medium',
    'high': 'High', 'HIGH': 'High', 'H': 'High', 'hi': 'High'
}
if 'stress_level' in df_clean.columns:
    df_clean['stress_level'] = df_clean['stress_level'].map(stress_mapping).fillna(df_clean['stress_level'])
    print(f"Stress level standardized: {df_clean['stress_level'].unique()}")

impact_mapping = {
    'no': 'No', 'NO': 'No', 'N': 'No', 'n': 'No', 'Nope': 'No',
    'yes': 'Yes', 'YES': 'Yes', 'Y': 'Yes', 'y': 'Yes', 'Yeah': 'Yes'
}
if 'academic_work_impact' in df_clean.columns:
    df_clean['academic_work_impact'] = df_clean['academic_work_impact'].map(impact_mapping).fillna(df_clean['academic_work_impact'])
    print(f"Academic work impact standardized: {df_clean['academic_work_impact'].unique()}")

addiction_mapping = {
    'none': 'None', 'NONE': 'None',
    'mild': 'Mild', 'MILD': 'Mild',
    'moderate': 'Moderate', 'MOD': 'Moderate',
    'severe': 'Severe', 'SEV': 'Severe',
    'high': 'Moderate', 'extreme': 'Severe'
}
if 'addiction_level' in df_clean.columns:
    df_clean['addiction_level'] = df_clean['addiction_level'].map(addiction_mapping).fillna(df_clean['addiction_level'])
    print(f"Addiction level standardized: {df_clean['addiction_level'].unique()}")

# ============================================================
# STEP 5: Fix Label Inconsistencies
# ============================================================
print("\n--- Step 5: Fix Label Inconsistencies ---")

if 'addicted_label' in df_clean.columns and 'addiction_level' in df_clean.columns:
    mismatch_0 = (df_clean['addicted_label'] == 0) & (df_clean['addiction_level'].isin(['Moderate', 'Severe']))
    df_clean.loc[mismatch_0, 'addicted_label'] = 1
    print(f"Fixed {mismatch_0.sum()} records: label 0->1 where level=Moderate/Severe")

    mismatch_1 = (df_clean['addicted_label'] == 1) & (df_clean['addiction_level'].isin(['None', 'Mild']))
    df_clean.loc[mismatch_1, 'addicted_label'] = 0
    print(f"Fixed {mismatch_1.sum()} records: label 1->0 where level=None/Mild")

# ============================================================
# STEP 6: Handle Remaining Missing Values
# ============================================================
print("\n--- Step 6: Impute Remaining Missing Values ---")

numeric_impute_cols = ['age', 'daily_screen_time_hours', 'social_media_hours', 'gaming_hours',
                       'work_study_hours', 'sleep_hours', 'notifications_per_day',
                       'app_opens_per_day', 'weekend_screen_time']

for col in numeric_impute_cols:
    if col in df_clean.columns and df_clean[col].isna().sum() > 0:
        median_val = df_clean[col].median()
        na_count = df_clean[col].isna().sum()
        df_clean[col].fillna(median_val, inplace=True)
        print(f"  {col}: Imputed {na_count} values with median={median_val:.2f}")

categorical_impute_cols = ['gender', 'stress_level', 'academic_work_impact', 'addiction_level']
for col in categorical_impute_cols:
    if col in df_clean.columns and df_clean[col].isna().sum() > 0:
        mode_val = df_clean[col].mode()[0]
        na_count = df_clean[col].isna().sum()
        df_clean[col].fillna(mode_val, inplace=True)
        print(f"  {col}: Imputed {na_count} values with mode={mode_val}")

print(f"\nData after cleaning: {df_clean.shape[0]} records, {df_clean.shape[1]} columns")
print(f"Remaining missing values: {df_clean.isna().sum().sum()}")
print(f"Removed {len(df) - df_clean.shape[0]} records total (duplicates)")
```

```python
**Interpretation of Data Cleaning Steps:**

The dataset contained several realistic data quality issues that required careful handling:

1. **Duplicates**: ~1.5% of records were exact duplicates, likely from data entry errors or system glitches. Removing them prevents bias toward duplicated patterns.

2. **Data Type Issues**: Some ages were stored as strings instead of numbers, which would cause errors in modeling. `pd.to_numeric()` with `errors='coerce'` converts valid strings and sets invalid ones to NaN.

3. **Unrealistic Values**: 
   - Ages below 13 or above 100 are unrealistic for smartphone usage studies
   - Screen time exceeding 24 hours is physically impossible
   - Negative values for hours/notifications are data entry errors
   - These were set to NaN rather than deleted to preserve other valid features in those rows

4. **Categorical Standardization**: Multiple variations of the same category (e.g., 'male', 'MALE', 'M', 'm') were mapped to a single standard form ('Male'). This is critical because ML models would treat these as different categories.

5. **Label Inconsistencies**: Records where `addicted_label` contradicted `addiction_level` were corrected based on the more detailed `addiction_level` field, which provides finer-grained information.

6. **Missing Value Imputation**: After cleaning, remaining NaN values were imputed using median (for numeric - robust to outliers) and mode (for categorical - most frequent value).
```

### 3.2 Feature Encoding and Transformation

```python
# Encode categorical variables
le = LabelEncoder()

df_encoded = df_clean.copy()

# Encode gender
df_encoded['gender_encoded'] = le.fit_transform(df_encoded['gender'])

# Encode stress_level (ordinal: Low=0, Medium=1, High=2)
stress_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
df_encoded['stress_level_encoded'] = df_encoded['stress_level'].map(stress_mapping)

# Encode academic_work_impact (ordinal: No=0, Yes=1)
impact_mapping = {'No': 0, 'Yes': 1}
df_encoded['academic_work_impact_encoded'] = df_encoded['academic_work_impact'].map(impact_mapping)

# Encode addiction_level (ordinal: None=0, Mild=1, Moderate=2, Severe=3)
addiction_mapping = {'None': 0, 'Mild': 1, 'Moderate': 2, 'Severe': 3}
df_encoded['addiction_level_encoded'] = df_encoded['addiction_level'].map(addiction_mapping)

# Verify no NaN in encoded columns
print("Encoding completed!")
print(f"Encoded dataset shape: {df_encoded.shape}")
print(f"NaN check in encoded columns:")
for col in ['gender_encoded', 'stress_level_encoded', 'academic_work_impact_encoded', 'addiction_level_encoded']:
    nan_count = df_encoded[col].isna().sum()
    if nan_count > 0:
        print(f"  WARNING: {col} has {nan_count} NaN values - filling with 0")
        df_encoded[col].fillna(0, inplace=True)
    else:
        print(f"  {col}: OK")
```

## 5. Feature Engineering

### 5.1 Creating New Features

**Interpretation**: Feature engineering helps create more informative variables that can improve model performance.

```python
# Create new features based on domain knowledge
df_features = df_encoded.copy()

# 1. Screen time ratio (weekend vs weekday)
df_features['screen_time_ratio'] = df_features['weekend_screen_time'] / (df_features['daily_screen_time_hours'] + 1e-6)

# 2. Social media engagement ratio
df_features['social_media_ratio'] = df_features['social_media_hours'] / (df_features['daily_screen_time_hours'] + 1e-6)

# 3. Gaming engagement ratio
df_features['gaming_ratio'] = df_features['gaming_hours'] / (df_features['daily_screen_time_hours'] + 1e-6)

# 4. Notification intensity (notifications per hour of screen time)
df_features['notification_intensity'] = df_features['notifications_per_day'] / (df_features['daily_screen_time_hours'] + 1e-6)

# 5. App usage frequency (app opens per hour of screen time)
df_features['app_frequency'] = df_features['app_opens_per_day'] / (df_features['daily_screen_time_hours'] + 1e-6)

# 6. Sleep quality indicator (inverse of sleep hours, normalized)
df_features['sleep_quality'] = 8 - df_features['sleep_hours']  # 8 hours is considered optimal
df_features['sleep_quality'] = df_features['sleep_quality'].clip(lower=0)

# 7. Digital wellness score (combination of positive factors)
df_features['digital_wellness'] = (
    (8 - df_features['sleep_hours']) * 0.3 +  # Less sleep = lower wellness
    df_features['daily_screen_time_hours'] * 0.2 +  # More screen time = lower wellness
    df_features['notifications_per_day'] / 100 * 0.2 +  # More notifications = lower wellness
    df_features['stress_level_encoded'] * 0.3  # Higher stress = lower wellness
)

# 8. Age-risk interaction
df_features['age_risk'] = df_features['age'] * df_features['daily_screen_time_hours'] / 100

# 9. Total digital activities
df_features['total_digital_activities'] = (
    df_features['social_media_hours'] + 
    df_features['gaming_hours'] + 
    df_features['work_study_hours']
)

# 10. Work-life balance indicator
df_features['work_life_balance'] = df_features['work_study_hours'] / (df_features['daily_screen_time_hours'] + 1e-6)

print("Feature engineering completed!")
print(f"New dataset shape: {df_features.shape}")
print("\nNew features created:")
new_features = ['screen_time_ratio', 'social_media_ratio', 'gaming_ratio', 
               'notification_intensity', 'app_frequency', 'sleep_quality',
               'digital_wellness', 'age_risk', 'total_digital_activities', 'work_life_balance']
for feature in new_features:
    if feature in df_features.columns:
        print(f"- {feature}: {df_features[feature].describe()}")
```

### 6.2 Model Training and Comparison

**Interpretation**: We'll train multiple algorithms to find the best performing model for our smartphone addiction detection task.

```python
# Define models to compare
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(random_state=42, eval_metric='logloss'),
    'LightGBM': lgb.LGBMClassifier(random_state=42, verbose=-1),
    'SVM': SVC(random_state=42, probability=True),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'Neural Network': MLPClassifier(random_state=42, max_iter=1000, hidden_layer_sizes=(100, 50))
}

# Train and evaluate models
results = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Training and evaluating models...")
print("=" * 60)

for name, model in models.items():
    try:
        # Cross-validation scores
        cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='accuracy')
        
        # Train on full training set
        model.fit(X_train_scaled, y_train)
        
        # Predictions
        y_pred = model.predict(X_test_scaled)
        y_pred_proba = model.predict_proba(X_test_scaled)[:, 1] if hasattr(model, 'predict_proba') else None
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None
        
        results[name] = {
            'model': model,
            'cv_mean': cv_scores.mean(),
            'cv_std': cv_scores.std(),
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'auc': auc
        }
        
        print(f"{name:20s}: CV={cv_scores.mean():.4f}±{cv_scores.std():.4f}, "
              f"Acc={accuracy:.4f}, F1={f1:.4f}, AUC={auc:.4f if auc else 'N/A':>6s}")
        
    except Exception as e:
        print(f"{name:20s}: Error - {str(e)}")
        results[name] = None
```

### 6.3 Hyperparameter Tuning for Best Models

**Interpretation**: We'll tune the top 3 performing models to optimize their performance.

```python
# Select top 3 models for hyperparameter tuning
top_models = sorted(valid_results.items(), key=lambda x: x[1]['f1'], reverse=True)[:3]
print("Top 3 models selected for hyperparameter tuning:")
for i, (name, result) in enumerate(top_models, 1):
    print(f"{i}. {name} (F1: {result['f1']:.4f})")

# Define parameter grids for tuning
param_grids = {
    'Random Forest': {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 6, 10],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 0.9, 1.0]
    },
    'LightGBM': {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 6, 10],
        'learning_rate': [0.01, 0.1, 0.2],
        'num_leaves': [31, 50, 100]
    }
}

# Perform hyperparameter tuning
tuned_results = {}

for model_name, _ in top_models:
    if model_name in param_grids:
        print(f"\nTuning {model_name}...")
        
        # Get the base model
        if model_name == 'Random Forest':
            base_model = RandomForestClassifier(random_state=42)
        elif model_name == 'XGBoost':
            base_model = xgb.XGBClassifier(random_state=42, eval_metric='logloss')
        elif model_name == 'LightGBM':
            base_model = lgb.LGBMClassifier(random_state=42, verbose=-1)
        
        # Grid search with cross-validation
        grid_search = GridSearchCV(
            base_model, 
            param_grids[model_name], 
            cv=3, 
            scoring='f1', 
            n_jobs=-1, 
            verbose=1
        )
        
        grid_search.fit(X_train_scaled, y_train)
        
        # Evaluate best model
        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_test_scaled)
        y_pred_proba = best_model.predict_proba(X_test_scaled)[:, 1]
        
        tuned_results[model_name] = {
            'best_model': best_model,
            'best_params': grid_search.best_params_,
            'best_score': grid_search.best_score_,
            'test_accuracy': accuracy_score(y_test, y_pred),
            'test_f1': f1_score(y_test, y_pred),
            'test_auc': roc_auc_score(y_test, y_pred_proba)
        }
        
        print(f"Best F1 Score: {grid_search.best_score_:.4f}")
        print(f"Test F1 Score: {tuned_results[model_name]['test_f1']:.4f}")
        print(f"Best Parameters: {grid_search.best_params_}")
```

### 8.2 Create Prediction Function for Deployment

```python
def predict_smartphone_addiction(user_data, pipeline_path='smartphone_addiction_pipeline.pkl'):
    """
    Predict smartphone addiction based on user data.
    
    Parameters:
    -----------
    user_data : dict
        Dictionary containing user features. Expected keys:
        - age: int
        - gender: str ('Male', 'Female', 'Other')
        - daily_screen_time_hours: float
        - social_media_hours: float
        - gaming_hours: float
        - work_study_hours: float
        - sleep_hours: float
        - notifications_per_day: int
        - app_opens_per_day: int
        - weekend_screen_time: float
        - stress_level: str ('Low', 'Medium', 'High')
        - academic_work_impact: str ('No', 'Yes')
    
    pipeline_path : str
        Path to the saved pipeline file
    
    Returns:
    --------
    dict
        Prediction results with probability and risk level
    """
    
    # Load pipeline
    pipeline = joblib.load(pipeline_path)
    model = pipeline['model']
    scaler = pipeline['scaler']
    features = pipeline['features']
    metadata = pipeline['metadata']
    
    # Create feature vector
    feature_vector = {}
    
    # Basic features
    feature_vector['age'] = user_data['age']
    feature_vector['daily_screen_time_hours'] = user_data['daily_screen_time_hours']
    feature_vector['social_media_hours'] = user_data['social_media_hours']
    feature_vector['gaming_hours'] = user_data['gaming_hours']
    feature_vector['work_study_hours'] = user_data['work_study_hours']
    feature_vector['sleep_hours'] = user_data['sleep_hours']
    feature_vector['notifications_per_day'] = user_data['notifications_per_day']
    feature_vector['app_opens_per_day'] = user_data['app_opens_per_day']
    feature_vector['weekend_screen_time'] = user_data['weekend_screen_time']
    
    # Encoded categorical features
    feature_vector['gender_encoded'] = metadata['feature_encodings']['gender'].get(user_data['gender'], 2)
    feature_vector['stress_level_encoded'] = metadata['feature_encodings']['stress_level'].get(user_data['stress_level'], 1)
    feature_vector['academic_work_impact_encoded'] = metadata['feature_encodings']['academic_work_impact'].get(user_data['academic_work_impact'], 0)
    
    # Engineered features
    feature_vector['screen_time_ratio'] = user_data['weekend_screen_time'] / (user_data['daily_screen_time_hours'] + 1e-6)
    feature_vector['social_media_ratio'] = user_data['social_media_hours'] / (user_data['daily_screen_time_hours'] + 1e-6)
    feature_vector['gaming_ratio'] = user_data['gaming_hours'] / (user_data['daily_screen_time_hours'] + 1e-6)
    feature_vector['notification_intensity'] = user_data['notifications_per_day'] / (user_data['daily_screen_time_hours'] + 1e-6)
    feature_vector['app_frequency'] = user_data['app_opens_per_day'] / (user_data['daily_screen_time_hours'] + 1e-6)
    feature_vector['sleep_quality'] = max(0, 8 - user_data['sleep_hours'])
    feature_vector['digital_wellness'] = (
        (8 - user_data['sleep_hours']) * 0.3 +
        user_data['daily_screen_time_hours'] * 0.2 +
        user_data['notifications_per_day'] / 100 * 0.2 +
        feature_vector['stress_level_encoded'] * 0.3
    )
    feature_vector['age_risk'] = user_data['age'] * user_data['daily_screen_time_hours'] / 100
    feature_vector['total_digital_activities'] = (
        user_data['social_media_hours'] + 
        user_data['gaming_hours'] + 
        user_data['work_study_hours']
    )
    feature_vector['work_life_balance'] = user_data['work_study_hours'] / (user_data['daily_screen_time_hours'] + 1e-6)
    
    # Create feature array in the correct order
    feature_array = np.array([feature_vector[feature] for feature in features]).reshape(1, -1)
    
    # Scale features
    feature_array_scaled = scaler.transform(feature_array)
    
    # Make prediction
    prediction_proba = model.predict_proba(feature_array_scaled)[0][1]
    prediction = int(prediction_proba >= metadata['best_threshold'])
    
    # Determine risk level
    if prediction_proba < 0.3:
        risk_level = 'LOW'
    elif prediction_proba < 0.6:
        risk_level = 'MEDIUM'
    elif prediction_proba < 0.8:
        risk_level = 'HIGH'
    else:
        risk_level = 'CRITICAL'
    
    return {
        'prediction': prediction,
        'probability': float(prediction_proba),
        'risk_level': risk_level,
        'confidence': float(max(prediction_proba, 1 - prediction_proba) * 100),
        'model_used': metadata['model_name'],
        'threshold_used': metadata['best_threshold']
    }

# Test the prediction function
test_user = {
    'age': 25,
    'gender': 'Male',
    'daily_screen_time_hours': 8.5,
    'social_media_hours': 4.2,
    'gaming_hours': 2.1,
    'work_study_hours': 3.5,
    'sleep_hours': 6.0,
    'notifications_per_day': 150,
    'app_opens_per_day': 80,
    'weekend_screen_time': 10.2,
    'stress_level': 'High',
    'academic_work_impact': 'Yes'
}

prediction_result = predict_smartphone_addiction(test_user)
print("Sample Prediction:")
print(json.dumps(prediction_result, indent=2))
```
